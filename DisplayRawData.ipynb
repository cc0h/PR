```python
import os
import numpy as np
from PIL import Image

# 定义文件路径
input_dir = 'face/rawdata'
output_dir = 'face/processedData'

# 创建输出目录（如果不存在）
os.makedirs(output_dir, exist_ok=True)

# 遍历输入目录中的所有文件
for filename in os.listdir(input_dir):
        file_path = os.path.join(input_dir, filename)
        file_size_kb = os.path.getsize(file_path) / 1024
        # 读取二进制文件
        with open(file_path, 'rb') as f:
            data = np.fromfile(f, dtype=np.uint8)
            # 将数据重塑为 128x128 或 512x512 的矩阵
        if file_size_kb <20:
            image = data.reshape(128, 128)
        else:
            image = data.reshape(512, 512)

        # 将处理后的图像转换为 Pillow 图像对象
        processed_image = Image.fromarray(image)

        # 构建输出文件路径
        output_file_path = os.path.join(output_dir, f'{filename}.png')

        # 保存处理后的图像
        processed_image.save(output_file_path)

print("图像处理完成并保存到 face/processedData 目录")



```
